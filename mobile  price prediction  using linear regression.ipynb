# splitting the data into train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
#for .25=92% , .30=91% and .35=91%
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred,y_test)
```

# RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred= rf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

# Ridge Regression

```python
from sklearn.model_selection import GridSearchCV #to get best set of hyperparameters
from sklearn.linear_model import Ridge

ridge = Ridge()

parameters = {'alpha': [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20,30]}

ridge_regressor = GridSearchCV(ridge, parameters)

ridge_regressor.fit(x_train, y_train)
```

```python
ridge_regressor.best_params_ #passes all combinations of hyperparameters one by one into the model and check the result
```

```python
ridge=Ridge(alpha=10)
ridge.fit(x_train,y_train)
y_pred=ridge.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

# Elastic Net

```python
from sklearn.linear_model import ElasticNet

elastic = ElasticNet()

parameters = {'alpha': [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20]}

elastic_regressor = GridSearchCV(elastic, parameters)

elastic_regressor.fit(x_train, y_train)
```

```python
elastic_regressor.best_params_
```

```python
elastic=ElasticNet(alpha=.01)
elastic.fit(x_train,y_train)
y_pred=elastic.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
